# Fetal Health — Correlation Matrix
This notebook loads `Fetal-health/dataset/fetal_health.csv`, computes the correlation matrix for the numeric features, displays it, and plots a heatmap. Outputs are saved into the dataset folder.

```python
%matplotlib inline
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paths
csv_path = os.path.join('dataset', 'fetal_health.csv')
heatmap_path = os.path.join('dataset', 'correlation_heatmap.png')

# Load dataset (semicolon-separated)
df = pd.read_csv(csv_path, sep=';')
print('Data shape:', df.shape)
display(df.head())

# Compute correlation matrix on numeric columns
corr = df.corr()
display(corr.round(3))

# Plot and save heatmap
plt.figure(figsize=(26,18))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Feature correlation heatmap')
plt.tight_layout()
plt.savefig(heatmap_path, dpi=150)
plt.show()
print('Saved heatmap to', heatmap_path)
```
